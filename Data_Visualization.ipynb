```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('C:/Users/Administrator/Desktop/birds.csv')
birds.head()
```

# Scatter plot

```python
birds.plot(kind='scatter',x='MaxLength',y='Order',figsize=(12,8))

plt.title('Max Length per Order')
plt.ylabel('Order')
plt.xlabel('Max Length')

plt.show()
```

# Hist plot

```python
birds['MaxBodyMass'].plot(kind = 'hist', bins = 10, figsize = (12,6))
plt.show()
```

# Density plot

```python
birds['MaxBodyMass'].plot(kind = 'density', figsize = (12,6))
plt.show()
```

# Line plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.relplot(kind="line", data=birds, x="MaxLength", y="MaxBodyMass", hue="Order", ci="sd", palette="viridis", height=6, aspect=2);
```

# Bar Chart plots

```python
birds['MaxBodyMass'].plot(kind='bar', figsize=(12, 6))
plt.show()
```
